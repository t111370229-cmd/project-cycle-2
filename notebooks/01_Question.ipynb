# 01_研究問題

本筆記本的目標有兩個：

1. 固定本專題的兩個研究問題  
2. 從原始資料表中擷取 `SadOrHopeless` 與 `HowMuchDoYouWeighWithoutShoesInKG` 兩個變數，輸出只包含這兩欄的新資料表


## 1. Research Questions

### Proportion Analysis
Is the proportion of students who felt sad or hopeless different from 0.30?

**中文：**  
YRBS 2007 資料中，感到 sad or hopeless 的學生比例是否與 0.30 不同？

### Mean Analysis
Is the mean weight of students different from 68.0 kg?

**中文：**  
YRBS 2007 資料中，學生平均體重是否與 68.0 公斤不同？


```python
# 匯入套件
import pandas as pd

# 設定顯示格式
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

# 讀取原始資料
df = pd.read_csv("../data/raw/YRBS_2007.csv")

# 顯示原始資料基本資訊
print("原始資料維度：", df.shape)
print("\n原始欄位名稱：")
print(df.columns.tolist())
```

```python
# 指定本專題要使用的兩個變數
selected_columns = [
    "SadOrHopeless",
    "HowMuchDoYouWeighWithoutShoesInKG"
]

# 擷取指定欄位
df_selected = df[selected_columns].copy()

# 顯示前幾列
print("擷取後資料維度：", df_selected.shape)
print("\n前 5 列資料：")
display(df_selected.head())
```

```python
# 建立輸出路徑並輸出只含兩個變數的新資料表
output_path = "../data/processed/cycle2_selected_variables.csv"
df_selected.to_csv(output_path, index=False)

print(f"已輸出檔案：{output_path}")
```

```python
# 再次確認輸出的資料表欄位
print("輸出資料表欄位：")
print(df_selected.columns.tolist())
```

## 2. Output

本筆記本輸出的檔案為：

- `../data/processed/cycle2_selected_variables.csv`

此檔案只包含後續分析需要的兩個變數：
- `SadOrHopeless`
- `HowMuchDoYouWeighWithoutShoesInKG`
